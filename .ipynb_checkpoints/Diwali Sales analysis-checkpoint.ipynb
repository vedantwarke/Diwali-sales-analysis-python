```python
#Stacked Bar Chart 
x=np.array(["0-17","18-25","26-35","36-45","46-50","51-55","55+"])
y=np.array([162,1305,3269,1578,693,553,272])
z=np.array([134,574,1272,705,290,277,155])
plt.bar(x,y,label="Female")
plt.bar(x,z,bottom=y, label="Male")
plt.legend(title="Gender")

plt.show()
```

```python
#MULTIPLE BAR CHARTS

width=0.4

x=np.array(["0-17","18-25","26-35","36-45","46-50","51-55","55+"])
y=np.array([162,1305,3269,1578,693,553,272])
z=np.array([134,574,1272,705,290,277,155])

values=np.arange(len(x))
print(values)
plt.bar(x,y,width, label="Female")
plt.bar(values+width,z,width, label="Male")
plt.legend(title="Gender")

plt.show()
```

```python
df.groupby('Age Group')['Amount'].sum()
```

```python
#Amount spent acd to Age group 
grouped_data=df.groupby('Age Group')['Amount'].sum()
col=['red','blue','green','hotpink','cyan','yellow','orange']
plt.bar(grouped_data.index, grouped_data.values,color=col)
```

```python
# Analysis of Orders State Wise 
gd=df.groupby('State')['Orders'].sum().head(10)
plt.figure(figsize=(15, 6))
col=['red','blue','green','hotpink','cyan','yellow','orange','lime','grey','violet']
plt.bar(gd.index,gd.values,color=col)
plt.show()
```

```python
df
```

```python
#Marital Status Count
x=np.array([0,1])
gd=df.groupby('Marital_Status')['Marital_Status'].count()
plt.bar(x,gd.values)
for i in range(len(x)):
    plt.text(i, gd[i]+100, gd[i],ha="center", va="center")
plt.show()
```

```python
#Occupation v/s Count
gd=df.groupby('Occupation')['User_ID'].count()
plt.figure(figsize=(25, 6))
col=['red','blue','green','hotpink','cyan','yellow','orange','lime','grey','violet']
plt.bar(gd.index,gd.values,color=col)
plt.show()
```

```python
#Occupation v/s Amount
gd=df.groupby('Occupation')['Amount'].sum()
plt.figure(figsize=(25, 6))
col=['red','blue','green','hotpink','cyan','yellow','orange','lime','grey','violet']
plt.bar(gd.index,gd.values,color=col)
plt.show()
```

```python
#Product Category v/s Count
gd=df.groupby('Product_Category')['User_ID'].count()
plt.figure(figsize=(25, 6))
col=['red','blue','green','hotpink','cyan','yellow','orange','lime','grey','violet']
plt.bar(gd.index,gd.values,color=col)
plt.show()
```

```python
#Product Category v/s Amount
gd=df.groupby('Product_Category')['Amount'].sum()
plt.figure(figsize=(25, 6))
col=['red','blue','green','hotpink','cyan','yellow','orange','lime','grey','violet']
plt.bar(gd.index,gd.values,color=col)
plt.show()
```

```python
#Product Orders

gd=df.groupby('Product_ID')['Orders'].count().sort_values(ascending=False).head(10)
plt.figure(figsize=(25, 6))
col=['red','blue','green','hotpink','cyan','yellow','orange','lime','grey','violet']
plt.bar(gd.index,gd.values,color=col)
plt.show()
```
